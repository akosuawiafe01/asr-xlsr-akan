```python
import re
chars_to_ignore_regex = '[\,\?\.\!\-\;\:\"\“\%\‘\”\…\)\]\’\[\\\/\�]'

def remove_special_characters(batch):
    batch["Transcriptions"] = re.sub(chars_to_ignore_regex, '', batch["Transcriptions"]).lower() + " "
    return batch
```

```python
akan_voice_train_1 = akan_voice_train_1.map(remove_special_characters)
akan_voice_train_2= akan_voice_train_2.map(remove_special_characters)

akan_voice_test = akan_voice_test.map(remove_special_characters)
```

```python
import random
from IPython.display import display, HTML

def show_random_elements(dataset, num_examples=10):
    assert num_examples <= len(dataset), "Can't pick more elements than there are in the dataset."
    picks = []
    for _ in range(num_examples):
        pick = random.randint(0, len(dataset)-1)
        while pick in picks:
            pick = random.randint(0, len(dataset)-1)
        picks.append(pick)

    df = pd.DataFrame(dataset[picks])
    display(HTML(df.to_html()))
```

```python
show_random_elements(akan_voice_train_1.remove_columns(["wav_Full_Filename", 'Full Filename']), num_examples=20)
```

```python
def replace_weird_characters(batch):
  batch["Transcriptions"] = re.sub('[ℇ]', 'ɛ', batch["Transcriptions"])
  batch["Transcriptions"] = re.sub('[ↄ]', 'ɔ', batch["Transcriptions"])
  return batch
```

```python
akan_voice_train_1 = akan_voice_train_1.map(replace_weird_characters)
akan_voice_train_2 = akan_voice_train_2.map(replace_weird_characters)
akan_voice_test = akan_voice_test.map(replace_weird_characters)
```

```python
def extract_all_chars(batch):
  all_text = " ".join(batch["Transcriptions"])
  vocab = list(set(all_text))
  return {"vocab": [vocab], "all_text": [all_text]}
```

```python
vocab_train_2 = akan_voice_train_2.map(extract_all_chars, batched=True, batch_size=-1, keep_in_memory=True, remove_columns=akan_voice_train_2.column_names)

vocab_train_1 = akan_voice_train_1.map(extract_all_chars, batched=True, batch_size=-1, keep_in_memory=True, remove_columns=akan_voice_train_1.column_names)
vocab_test = akan_voice_test.map(extract_all_chars, batched=True, batch_size=-1, keep_in_memory=True, remove_columns=akan_voice_test.column_names)
```

```python
vocab_list = list(set(vocab_train_2["vocab"][0]) |set(vocab_train_1["vocab"][0]) | set(vocab_test["vocab"][0]))
```

```python
vocab_dict = {v: k for k, v in enumerate(vocab_list)}

vocab_len = len(vocab_dict)
print(f"The length of our Akan Twi vocabulary is {vocab_len}")
```

```python
vocab_dict["[UNK]"] = len(vocab_dict)
vocab_dict["[PAD]"] = len(vocab_dict)

print(len(vocab_dict), '\n')

vocab_dict
```

## Creating an XLSR Model

```python
from transformers import Wav2Vec2CTCTokenizer,Wav2Vec2ForCTC,Wav2Vec2Processor,Trainer,TrainingArguments,Wav2Vec2FeatureExtractor
```

```python
tokenizer = Wav2Vec2CTCTokenizer(f"{model_output_path}/vocab.json", unk_token="[UNK]", pad_token="[PAD]", word_delimiter_token="|")
```

```python
feature_extractor = Wav2Vec2FeatureExtractor(feature_size=1, sampling_rate=16000, padding_value=0.0, do_normalize=True, return_attention_mask=True)
```

```python
processor = Wav2Vec2Processor(feature_extractor=feature_extractor, tokenizer=tokenizer)
```

```python
processor.save_pretrained(f"{model_output_path}/wav2vec2-large-xlsr-AKAN")
```

```python
import torchaudio

def speech_file_to_array_fn(batch):
  # Loads audio files and creates corresponding speech 1D array, sampling_rate and transcription
  speech_array, sampling_rate = torchaudio.load(batch['full_wav_path'])
  batch['speech'] = speech_array[0].numpy()
  batch["sampling_rate"] = sampling_rate

  if batch["Transcriptions"]:
    batch["target_text"] = batch["Transcriptions"]
  else:
    batch["target_text"] = ""


  return batch
```

```python
def prepare_dataset(batch):
    # check that all files have the correct sampling rate
    # assert (
    #     len(set(batch["sampling_rate"])) == 1
    # ), f"Make sure all inputs have the same sampling rate of {processor.feature_extractor.sampling_rate}."

  batch["input_values"] = processor(batch["speech"], sampling_rate=batch["sampling_rate"][0]).input_values

  with processor.as_target_processor():
    batch["labels"] = processor(batch["target_text"]).input_ids
  return batch
```
