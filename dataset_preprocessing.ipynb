```python
def read_sr(audio_file):
  signal, sr = sf.read(audio_file)
  # print(sr)

  return sr
```

```python
def sample_rate(filename):
  path = f"{audio_path}/{filename}"

  save_path = f"{sr_audio_path}"

  # print(path, save_path)

  if not os.path.exists(save_path):
    os.makedirs(save_path, exist_ok=True)


  if os.path.exists(path):
    try:
      sound = AudioSegment.from_mp3(path)
      sound = sound.set_frame_rate(16000)
      sound.export(f"{save_path}/{filename[:-4]}.wav", format="wav")
    except:
      print(path)
```

```python
# %%capture
audio_mp3_files = os.listdir(audio_path)

# audio_mp3_files = audio_files

Parallel(n_jobs=8, backend="multiprocessing")(
    delayed(sample_rate)(filename) for filename in tqdm(audio_mp3_files)
)
```
